```python
from sklearn.datasets import load_iris
import pandas as pd

# Load the Iris dataset
imported = load_iris()

# Create a Pandas DataFrame
iris_df = pd.DataFrame(imported.data, columns=imported.feature_names)

# Add the target variable to the DataFrame
#iris_df['target'] = imported.target

# Print the first 5 rows of the DataFrame
iris_df.head()
```

```python
# Features (X) and Target (Y)
X = iris_df[["petal length (cm)" , "sepal width (cm)" , "petal width (cm)"]]
Y = iris_df["sepal length (cm)"]
```

```python
#import necassary libraries
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
```

```python
# split data
X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=0.2, random_state=42)
```

```python
# model selection
mymodel = LinearRegression()

# Model Training
mymodel.fit(X_train, Y_train)

# model evaluation
Y_pred = mymodel.predict(X_test)

```

```python
# y = mx + c
m = mymodel.coef_
c = mymodel.intercept_
print(m)
print(c)
```

```python
# evaluate the model
mse = mean_squared_error(Y_test, Y_pred)
rmse = np.sqrt(mse)
r2 = r2_score(Y_test, Y_pred)

print("Mean Squared Error:", mse)
print("Root Mean Squared Error:", rmse)
print("R-squared:", r2)
```

```python
def corelation(dataset, threshold):
  col_corr = set
  cor_matrix = dataset.corr()
  for i in range(len(cor_matrix.columns)):
    for j in range(i):
      if abs(cor_matrix.iloc[i,j]) > threshold:
        colname = cor_matrix.columns[i]  # getting the name of column
        col_corr.add(colname)
  return col_corr
```
